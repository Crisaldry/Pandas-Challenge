```python
#import pandas
import pandas as pd
```

```python
#Load files 
school_data_path = "Resources/schools_complete.csv"
student_data_path ="Resources/students_complete.csv"

school_data_read = pd.read_csv(school_data_path)
student_data_read = pd.read_csv(student_data_path)

#merging the data into one file
school_data_complete = pd.merge(student_data_read, school_data_read, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

```python
#Total Number of unique Schools

school_count = len(school_data_read)
print(school_count)
```

```python
# Total students
total_students = len(student_data_read)
print(total_students)
```

```python
#Total Budget
total_budget = school_data_read['budget'].sum()
print(total_budget)

```

```python
#average math score
average_math_score = round(school_data_complete['math_score'].mean(),2) 
print(average_math_score) #reduce number of decimals to two

#percent of students passing #passing is 70% or 
passing_math_series = school_data_complete['math_score'] >= 70
num_passing_math = school_data_complete[passing_math_series].count()['student_name']
percent_passing_math = round((num_passing_math / total_students * 100),2)
percent_passing_math
```

```python
#average reading score
average_reading_score = round(school_data_complete['reading_score'].mean(),2) 
print(average_reading_score) #reduce number of decimals to two

#percent of students passing #passing is 70% or higher
passing_reading_series = (school_data_complete['reading_score'] >= 70)
num_passing_reading = school_data_complete[passing_reading_series].count()['student_name']
percent_passing_reading = round((num_passing_reading / total_students * 100),2)
print(percent_passing_reading)
```

```python
# overall passing both reading and math

overall_passing_total = school_data_complete[passing_reading_series & passing_math_series]\
    ['student_name'].count()
print(overall_passing_total)
overall_passing_percent = round(100 * overall_passing_total / total_students,2)
print(overall_passing_percent)
```

```python
#create dictionary 

district_summary_dict ={ 
    "Total Schools": [school_count], 
    "Total Students": [total_students],
    "Total Budget": [total_budget],
    "Average Math Score": [average_math_score],
    "Average Reading Score": [average_reading_score],
    "% Passing Math": [percent_passing_math],
    "% Passing Reading": [percent_passing_reading],
    "% Overall Passing": [overall_passing_percent]
    }

#district_summary_dict


#create a data frame for district summary 
district_summary_DF = pd.DataFrame(district_summary_dict)

district_summary_DF

```

```python
#School Summary: Average reading, math, and budget

avg_by_school = round(school_data_complete.groupby('school_name').mean(),2)
avg_columns = ["reading_score", "math_score", "budget"]
school_summary_DF = avg_by_school[avg_columns]
school_summary_DF

```

```python
students_per_school = school_data_complete.groupby('school_name').count()['Student ID']

school_summary_DF["students per school"] = students_per_school 

school_summary_DF
```

```python
#budget per student 

students_per_school = school_data_complete.groupby('school_name').count()['student_name']

school_summary_DF["students per school"] = students_per_school 

budget_per_students = school_summary_DF["budget"] / students_per_school

school_summary_DF["budget per students"] = budget_per_students 
 
school_summary_DF

```

```python
# Percent Passing 

# % passing math 
per_school_passing_math = school_data_complete[passing_math_series].groupby('school_name')['Student ID'].count()

per_school_passing_math

school_summary_DF["% Passing Math"] = per_school_passing_math

# % passing reading
per_school_passing_reading = school_data_complete[passing_reading_series].groupby('school_name')['Student ID'].count()

per_school_passing_reading

school_summary_DF["% Passing Reading"] = per_school_passing_reading

# % overall

overall_per_school_passing = school_data_complete[passing_math_series & passing_reading_series].groupby('school_name')['Student ID'].count()

school_summary_DF["percent Passing"] = overall_per_school_passing

school_summary_DF



```
